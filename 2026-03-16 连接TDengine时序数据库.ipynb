```python
import taos

def show_databases():
    conn = None
    cursor = None
    try:
        # 建立连接（使用你提供的参数）
        conn = taos.connect(host="192.168.17.49", 
                            user="root", 
                            password="Qwer!234", 
                            port=6030)
        cursor = conn.cursor()
        cursor.execute("SHOW DATABASES")
        
        # 获取所有结果
        rows = cursor.fetchall()
        print("=== 数据库列表 ===")
        for row in rows:
            print(row[0])  # 第一列是数据库名

    except Exception as e:
        print(f"操作失败: {e}")
    finally:
        if cursor:
            cursor.close()
        if conn:
            conn.close()

if __name__ == "__main__":
    show_databases()
```
