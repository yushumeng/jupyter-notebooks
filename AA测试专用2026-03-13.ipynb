```python
from pyspark.sql import functions as F

# 1. 获取两个表的唯一 request_id
node_ids = spark.table("dwd.dwd_oa_workflow_efficiency_nodes_f_1h").select("request_id").distinct()
result_ids = spark.table("dwd.dwd_oa_workflow_efficiency_result_f_1h").select("request_id").distinct()

# 2. 找出节点表有、结果表没有的 request_id
diff_ids = node_ids.join(result_ids, on="request_id", how="left_anti")
print(f"差异数量: {diff_ids.count()}")   # 应该输出 119

# 3. 关联节点表，查看这些流程的 workflow_name
diff_with_name = diff_ids.join(
    spark.table("dwd.dwd_oa_workflow_efficiency_nodes_f_1h"),
    on="request_id",
    how="left"
).select("request_id", "workflow_name").distinct()

print("节点表中差异流程的 workflow_name:")
diff_with_name.show(200, truncate=False)

# 4. 关联原始 request_base 表，查看 createdate 和 requestname
diff_with_base = diff_ids.join(
    spark.table("ods.ods_oa_workflow_requestbase_i_1h"),
    diff_ids.request_id == F.col("requestid"),
    "left"
).select(
    "request_id",
    F.col("createdate"),
    F.col("requestname"),
    F.col("status")
)

print("差异流程的原始信息:")
diff_with_base.show(200, truncate=False)
```
